```python
import random
import numpy as np
import math
import matplotlib.pyplot as plt
import matplotlib.font_manager
matplotlib.font_manager._rebuild()

#input experiment variables

print("Enter number of sides on each dice")
num_sides = int(input())
print("Enter maximum roll to decay, e.g. enter 2 if a 1 or 2 results in decay")
max_decay_roll = int(input())
decay_chance = max_decay_roll/num_sides
l_ambda = decay_chance
print(l_ambda)
```

```python
print("number of dice")
num_dice = int(input())
print("Enter number of rolls for each simulation")
num_rolls = int(input())
print("Enter number of simulations you wish to run")
sim_count = int(input())

N_final = []
t_half= []



fig, (ax1, ax2) = plt.subplots(1,2)
fig.suptitle('decaying dice')

for i in range(sim_count):#for each simulation
    N_t = [num_dice]
    N_theory=[]
    t_fin = 0
    for j in range(num_rolls):#repeating the rolling process within each simulation
        #print(j)
        n = N_t[j]
        #print(n)
        #generate random set of rolls for each trial
        roll = np.random.random(n)
        #count number of surviving atoms after roll
        num_alive = np.count_nonzero(roll>decay_chance)
        #print(np.count_nonzero(roll>max_decay_roll/max_sides))
        N_t = np.append(N_t, num_alive)
        
        N_theory = np.append(N_theory,num_dice*math.e**(-l_ambda*j))
        #if np.count_nonzero(roll>max_decay_roll/max_sides) <= (num_dice/2):
        #    if t_fin = 0:
        #        t_half = np.append(t_half, j+1)
        #        t_fin = 1       
    else:
        #print(N_t[j])#check final number of dice remaining after num_rolls
        N_final=np.append(N_final,N_t[j+1])
        ax1.plot(N_t)
        N_theory = np.append(N_theory,num_dice*math.e**(-l_ambda*(j+1)))
        #N_t=[]#probs not needed to clear count
        
else:
    #print(N_t)
    #print(N_final)
    #print(N_theory)
    #print(t_half)
    ax1.plot(N_theory,'.-')
    ax2.hist(N_final,align='left')

    
#add a count printout to each row in the 10 10 200 plot - number of simulations (relative frequency)    

```
